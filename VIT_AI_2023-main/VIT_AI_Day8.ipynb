## Outlier Detection

```python
# z score, percentile, iqr - removal methods
# replacement method- median
```

```python
sns.boxplot(df.bmi)
```

```python
## Outlier removal using IQR
```

```python
q1 = df.bmi.quantile(.25)
q3 = df.bmi.quantile(.75)
```

```python
q1
```

```python
q3
```

```python
IQR = q3 - q1
IQR
```

```python
upper_limit = q3+1.5*IQR
lower_limit = q1-1.5*IQR
```

```python
upper_limit
```

```python
df = df[df.bmi<upper_limit]
```

```python
sns.boxplot(df.bmi)
```

```python
df.shape
```

```python
## Percentile
```

```python
sns.boxplot(df.bmi)
```

```python
p99 = df.bmi.quantile(0.99)
p99
```

```python
df = df[df.bmi<=p99]
```

```python
sns.boxplot(df.bmi)
```

```python
# z score
```

```python
from scipy import stats
```

```python
bmi_zscore = stats.zscore(df.bmi)
```

```python
bmi_zscore
```

```python
df_z = df[np.abs(bmi_zscore)<=3]
```

```python
sns.boxplot(df_z.bmi)
```

```python
df.shape
```

```python
## Replacement of outliers using median
```

```python
sns.boxplot(df.bmi)
```

```python
q1 = df.bmi.quantile(.25)
q3 = df.bmi.quantile(.75)
```

```python
IQR = q3 - q1
IQR
```

```python
upper_limit = q3+1.5*IQR
lower_limit = q1-1.5*IQR
```

```python
df.median()
```

```python
df['bmi'] =np.where(df['bmi']>upper_limit,30,df['bmi'])
```

```python
sns.boxplot(df.bmi)
```
