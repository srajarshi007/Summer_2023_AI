```python
import pandas as pd
import numpy as np
```

```python
# Creating 2 series

s1 = pd.Series(np.random.rand(5))
s2 = pd.Series(np.random.rand(5))

print(s1,s2)
```

```python
# Creating dataframe with the series

df = pd.DataFrame([s1,s2])
df
```

```python
# Transpose - Coverting observations(rows) into features(column)

df = df.T
```

```python
df
```

```python
# Changing the column name

df.columns = ['S1','S2']
df
```

```python
# Creating dictionary

d = {'Name':['Hari', 'Srikanth', 'Saumya'],
     'Gender':['M','M','F']}

d
```

```python
df = pd.DataFrame(d)
df
```
